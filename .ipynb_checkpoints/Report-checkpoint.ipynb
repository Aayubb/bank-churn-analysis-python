# Churn analysis Report:

## Executive Summary

This report analyzes a dataset of 10,000 bank customers to identify the key drivers of customer churn. The analysis found that Germany has the highest churn rate at 32.44%, nearly double the overall average of 20.38%. To reduce churn, the bank should prioritize re-engaging inactive members and investigating the root cause of Germany's elevated churn rate.

## Overall Churn rate

The overall churn rate was as follows:

| Total Customers | Total Churned | Churn Rate |
| --- | --- | --- |
| 10000 | 2038 | 20.38% |

This churn rate means that roughly one in five customers (or subscribers) stopped doing business with our company over a specific time frame.


## Key Findings

#### Finding 1: Geography  

In Germany the churn rate is 32.44%, compared to the overall rate of 20.38%. Germany is almost double the average, suggesting the bank should investigate whether product offerings, pricing, or customer service standards differ in Germany compared to France and Spain.

#### Finding 2: Age

The age group (51-60) has the highest churn rate amongst the age groups (18-30),(31-40),(41-50),(60+). Suggesting making policies and changes 


#### Finding 3: Active Member Status

Regarding the customer's activity, non-active members churned at 26.87% and active members churned at 14.27%. Suggesting finding ways to bring back the non-active members before they churn.

#### Finding 4: Gender

Females(25%) tend to churn more than male(16%), suggesting the bank's products or communication style may not be equally serving female customers.


## Statistical tests

Geography, Gender and Activity were tested using a statistical test called the chi-square test and their p-values were less than 0.001  (p < 0.001) which means the pattern is real.

Age and Balance were tested using a t-test, both returning p < 0.001, confirming that the differences in means between churned and retained customers are statistically real. CreditScore also passed the t-test (p = 0.007), however the actual difference in means was only 6 points (645 vs 651), making it a weak practical predictor despite being statistically significant.


## Business Recommendations

1. Conduct a regional audit of Germany's product offerings, pricing, and customer service quality to identify why churn is nearly double that of France and Spain.

2. Design targeted retention programs for customers aged 51-60, such as loyalty rewards, dedicated relationship managers, or tailored financial products for pre-retirement needs.

3. Launch a re-engagement campaign targeting inactive members, such as personalised outreach, special offers, or check-in calls — given that inactive members churn at nearly twice the rate of active ones.


## Limitations

1. This analysis shows correlation, not causation, we know these groups churn more, but not definitively why.

2. The dataset is missing potentially important variables such as customer complaints history, product pricing, and competitor data

3. This is a descriptive analysis of past behaviour, it cannot predict which specific customers will churn next.
